```python
# Add type hints to a function that adds two integers and returns an integer.
def add(a: int, b: int) -> int:
    return a + b
```

```python
# Write a function with type hints that accepts a list of floats and returns their average as a float.
from typing import List
def avg(nums: List[float]) -> float:
    return sum(nums) / len(nums)
```

```python
# Create a function that accepts a dictionary of string keys and int values and returns a list of keys typed correctly.
from typing import Dict, List
def dict_keys(d: Dict[str, int]) -> List[str]:
    return list(d.keys())
```

```python
# Add type hints to a function that returns Optional[int] if the input string can be converted to int, otherwise None.
from typing import Optional
def to_int(s: str) -> Optional[int]:
    try: return int(s)
    except ValueError: return None
```

```python
# Write a function with a typed Tuple return: (min_value: int, max_value: int).
from typing import Tuple, List
def min_max(nums: List[int]) -> Tuple[int, int]:
    return min(nums), max(nums)
```

```python
# Use TypedDict to define a user profile structure and create two sample profiles.
from typing import TypedDict
class User(TypedDict):
    name: str
    age: int
    active: bool

u1: User = {"name": "A", "age": 20, "active": True}
u2: User = {"name": "B", "age": 30, "active": False}
```

```python
# Define a function with type hints that accepts *args of ints and returns their sum.
def sum_args(*args: int) -> int:
    return sum(args)
```

```python
# Use Union type hints to allow a function to return either a list or a dictionary based on a flag.
from typing import Union, List, Dict
def make(flag: bool) -> Union[List[int], Dict[str, int]]:
    return [1,2,3] if flag else {"a":1}
```

```python
# Create a function with **kwargs type hints that accepts named parameters for "name: str" and "active: bool".
from typing import Any
def show_user(**kwargs: Any) -> None:
    name: str = kwargs.get("name")
    active: bool = kwargs.get("active")
    print(name, active)
```

```python
# Annotate a function that takes a list of strings and returns a new list containing only alphabetic strings.
from typing import List
def only_alpha(words: List[str]) -> List[str]:
    return [w for w in words if w.isalpha()]
```

```python
# Write a function that takes a Callable[[int, int], int] as input and applies it to two numbers.
from typing import Callable
def apply(f: Callable[[int, int], int], a: int, b: int) -> int:
    return f(a, b)
```

```python
# Define a function using type hints that returns a Dict[str, List[int]] from grouped user IDs.
from typing import Dict, List
def group(ids: List[int]) -> Dict[str, List[int]]:
    return {"group": ids}
```
