```python
# Build a File Metadata Extractor that prints size, created time, modified time, and extension.
import os, sys, time
from pathlib import Path
p=Path(sys.argv[1]); s=p.stat()
print("size:",s.st_size); print("created:",time.ctime(s.st_ctime))
print("modified:",time.ctime(s.st_mtime)); print("extension:",p.suffix)
```

```python
# Create a Text Cleaner tool that removes extra spaces, tabs, and normalizes newline characters.
import sys, re
t=open(sys.argv[1]).read().replace("\r","").replace("\t"," ")
t=re.sub(" +"," ",t); print("\n".join(l.strip() for l in t.split("\n")))
```

```python
# Build a Terminal Notes App that lets the user add, list, search, and delete short text notes.
import sys; f="notes.txt"
c=sys.argv[1]
if c=="add": open(f,"a").write(" ".join(sys.argv[2:])+"\n")
elif c=="list": print(open(f).read())
elif c=="search": q=sys.argv[2].lower(); print("".join(l for l in open(f) if q in l.lower()))
elif c=="delete": i=int(sys.argv[2])-1; L=open(f).read().splitlines(); L.pop(i); open(f,"w").write("\n".join(L))
```

```python
# Make a Mini Log Formatter that takes raw lines and prepends timestamps to each line.
import sys, time
print("\n".join(f"{time.ctime()} {l.strip()}" for l in open(sys.argv[1])))
```

```python
# Create a Keyword Highlighter that highlights target words in uppercase inside any text file.
import sys, re
t=open(sys.argv[1]).read(); w=sys.argv[2]
print(re.sub(fr"\b{w}\b",w.upper(),t,flags=re.I))
```

```python
# Build a Directory Scanner that counts files by extension (.txt, .log, .json) and prints a summary.
import sys, os
from collections import Counter
c=Counter()
for r,_,fs in os.walk(sys.argv[1]):
    for f in fs: c[os.path.splitext(f)[1]]+=1
print(c)
```

```python
# Develop a Simple Encryption Tool using base64 to encode/decode strings from the CLI.
import sys,base64
m=sys.argv[1]; t=sys.argv[2]
print(base64.b64encode(t.encode()).decode() if m=="enc" else base64.b64decode(t.encode()).decode())
```

```python
# Make a Password Auditor that checks length, digits, symbols, and gives a rating (weak/medium/strong).
import sys,re
p=sys.argv[1]; s=0
s+=len(p)>=8; s+=bool(re.search(r"\d",p)); s+=bool(re.search(r"[A-Z]",p))
s+=bool(re.search(r"[a-z]",p)); s+=bool(re.search(r"\W",p))
print("strong" if s>=4 else "medium" if s>=2 else "weak")
```

```python
# Create a JSON Validator that reads a .json file and prints whether it is valid or invalid.
import sys,json
try: json.load(open(sys.argv[1])); print("valid")
except: print("invalid")
```

```python
# Build a Find-and-Replace Utility that replaces a word across a whole text file and saves a new version.
import sys
print(open(sys.argv[1]).read().replace(sys.argv[2],sys.argv[3]))
```

```python
# Make a Duplicate Line Remover that outputs the file with unique lines preserved in order.
import sys
s=set(); o=[]
for l in open(sys.argv[1]):
    if l not in s: s.add(l); o.append(l)
print("".join(o))
```

```python
# Develop a Filename Normalizer that strips illegal characters and replaces spaces with underscores.
import sys,re
print(re.sub(r"[^a-zA-Z0-9._]","_",sys.argv[1]))
```

```python
# Create a Text Summary Report that prints line count, word count, char count, and longest line length.
import sys
L=open(sys.argv[1]).read().splitlines()
print("lines:",len(L)); print("words:",sum(len(x.split()) for x in L))
print("chars:",sum(len(x) for x in L)); print("longest:",max((len(x) for x in L),default=0))
```

```python
# Build an Environment Variable Loader that reads a .env file and prints key/value pairs safely.
import sys
for l in open(sys.argv[1]):
    if "=" in l: k,v=l.strip().split("=",1); print(k,":",v)
```

```python
# Create a Search-in-Folder Utility that finds files containing a given keyword inside their contents.
import sys,os
root=sys.argv[1]; q=sys.argv[2].lower()
for r,_,fs in os.walk(root):
    for f in fs:
        p=os.path.join(r,f)
        if q in open(p,errors="ignore").read().lower(): print(p)
```
