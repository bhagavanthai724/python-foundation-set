```python
# Write a function that takes two numbers and returns their sum, difference, product, and quotient as a tuple.
def calc(a, b):
    return a+b, a-b, a*b, a/b
```

```python
# Create a function that accepts a list and returns a new list containing only the unique elements.
def unique(lst):
    return list(set(lst))
```

```python
# Write a function that takes a string and returns True if it is a palindrome.
def is_pal(s):
    s=s.lower().replace(" ","")
    return s==s[::-1]
```

```python
# Create a function that counts how many uppercase and lowercase letters are in a given sentence.
def count_cases(s):
    up=sum(c.isupper() for c in s)
    low=sum(c.islower() for c in s)
    return up, low
```

```python
# Write a function that accepts variable arguments (*args) and returns their total sum.
def total_sum(*args):
    return sum(args)
```

```python
# Create a function that accepts keyword arguments (**kwargs) and prints each key-value pair.
def show_kwargs(**kwargs):
    for k,v in kwargs.items():
        print(k, v)
```

```python
# Build a function that takes a filename and returns how many lines the file contains.
def count_lines(fname):
    return sum(1 for _ in open(fname))
```

```python
# Write a function that takes a list of numbers and returns only those greater than a threshold argument.
def above(lst, t):
    return [x for x in lst if x>t]
```

```python
# Create a function that accepts two lists and returns True if they have at least one common element.
def has_common(a, b):
    return bool(set(a) & set(b))
```

```python
# Write a function that takes a sentence and returns a dictionary mapping each word to its frequency.
def freq(sentence):
    words=sentence.split()
    return {w:words.count(w) for w in words}
```

```python
# Create a function that returns the largest even number in a list or None if no even number exists.
def largest_even(lst):
    ev=[x for x in lst if x%2==0]
    return max(ev) if ev else None
```

```python
# Write a function that takes a list of temperatures in Celsius and returns them in Fahrenheit.
def to_f(c_list):
    return [(c*9/5)+32 for c in c_list]
```

```python
# Build a function that formats a name into: "Last, First (Middle)" where middle is optional.
def fmt(first, last, middle=None):
    return f"{last}, {first} ({middle})" if middle else f"{last}, {first}"
```

```python
# Write a function that validates whether a string is a strong password using custom rules passed as arguments.
def strong_pw(pw, min_len=8, need_digit=True, need_upper=True, need_symbol=True):
    import re
    if len(pw)<min_len: return False
    if need_digit and not re.search(r"\d", pw): return False
    if need_upper and not re.search(r"[A-Z]", pw): return False
    if need_symbol and not re.search(r"\W", pw): return False
    return True
```

```python
# Create a function `safe_divide(a, b)` that returns a division result or the message "division not possible".
def safe_divide(a, b):
    return a/b if b!=0 else "division not possible"
```
