```python
# Build the core Log Analyzer function that takes a list of log lines (strings) and extracts structured fields such as timestamp, log level, and message, returning them in a normalized Python dictionary list (no file handling).
def analyze_logs(log_lines):
    results = []

    for line in log_lines:
        parts = line.split(maxsplit=2)
        if len(parts) == 3:
            timestamp, level, message = parts
        else:
            timestamp, level, message = None, None, line
        results.append({
            "timestamp": timestamp,
            "level": level,
            "message": message,
            "raw": line
        })
    return results
```
