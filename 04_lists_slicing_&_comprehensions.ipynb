```python
# Create a list of 10 integers and print only the even numbers using list comprehension.
lst = [1,2,3,4,5,6,7,8,9,0]
result = [i for i in lst if i % 2 == 0]
print(result)
```

```python
# Given a list of strings, convert all items to lowercase using list comprehension.
lst = ["AI" , "ML" , "DL" , "PYTHON"]
result = [i.lower() for i in lst]
print(result)
```

```python
# Slice a list to extract the first 3 elements, the last 3 elements, and every second element.
lst = ["1" , "2" , "3" , "4" , "5"]
print(lst[:3])
print(lst[-3:])
for i in range(len(lst)):
    if i % 2 == 0 and i != 0:
        print(lst[i],end = " ")
```

```python
# Write a script that removes all None values from a list using list comprehension.
lst = [1,2,3,4,5,6,None]
result = [i for i in lst if i is not None]
print(result)
```

```python
# Take a list of numbers and create a new list containing only those greater than the average.
lst = [1,2,3,4,5,6]
average = sum(lst)/len(lst)
lst_new = []
for i in lst:
    if i > average:
        lst_new.append(i)
print(lst_new)
```

```python
# Flatten a nested list (e.g., [[1,2],[3,4]]) into a single list using comprehension.
lst = [[1,2],[3,4],[5,6],[7,8]]
result = [j for i in lst for j in i]
print(result)
```

```python
# Reverse a list using slicing and also using the reversed() function.
lst = ["AI","Python","ML","DL","DS"]
result = list(reversed(lst))
print(result)
```

```python
# From a list of words, extract only those that start with a vowel using list comprehension.a
lst = ["ML","DL","GAI","ANI","AGI","ASI"]
result = [i for i in lst if i[0] in "aeiouAEIOU"]
print(result)
```

```python
# Given a list with duplicates, create a new list that contains only unique elements while preserving order.
lst = [1,2,2,3,4,4,5]
lst_new = []
for i in lst:
    if i not in lst_new:
        lst_new.append(i)
print(lst_new)
```

```python
# Use slicing to rotate a list by k positions to the right.
lst = [1, 2, 3, 4, 5]
n = 2
result = lst[-n:] + lst[:-n]
print(result)
```

```python
# Convert a list of integers to a list of strings using comprehension.
lst = [1,2,3,4,5]
result  = [str(i) for i in lst]
print(result)
```

```python
# Take a list of filenames and extract only those ending with ".log" or ".txt".
lst = ["AI.txt","Python.log","ML.txt","ANI","AGI","ASI"]
result  = [i for i in lst if i.endswith((".log" , ".txt"))]
print(result)
```

```python
# Build a list of squares for numbers from 1 to 20 but only include numbers divisible by 3.
lst = []
for i in range(1,21):
    if i % 3 == 0:
        print(i**2)
```

```python
# Given a list of temperatures in Celsius, convert them to Fahrenheit using comprehension.
lst = [10,20,30,40,50]
result = [(i*9/5)+32 for i in lst]
print(result)
```

```python
# Split a sentence into words and build a list containing the length of each word.
sentence = "Artificial intelligence is transforming the world"
words = sentence.split()
result = [len(word) for word in words]
print(words)
print(result)
```
