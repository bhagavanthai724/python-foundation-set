```python
# Write a program that takes a string and prints the first, middle, and last characters.
string = input("Enter the string : ")
if len(string) == 0:
    print("The string is empty. No characters to display.")
else:
  first = string[0]
  last = string[-1]
  middle = string[len(string)//2]
  print(f"First : {first}, Middle : {middle}, Last : {last}")
```

```python
# Given a string input, remove all leading and trailing spaces and print the cleaned version.
string = input("Enter the string : ").strip()
print(string)
```

```python
# Accept a sentence and replace all spaces with hyphens using string methods.
string = input("Enter the string : ")
result = string.replace(" ","-")
print(result)
```

```python
# Take a string and count how many digits, alphabets, and special characters it contains.
string = input("Enter the string: ")
digits = 0
alphabets = 0
special = 0
for ch in string:
    if ch.isdigit():
        digits += 1
    elif ch.isalpha():
        alphabets += 1
    else:
        special += 1
print("Alphabets:", alphabets)
print("Digits:", digits)
print("Special characters:", special)
```

```python
# Write a script to extract all words longer than 5 characters from a sentence.
string = input("Enter the string : ")
for i in string.split():
    if len(i) > 5:
        print(i)
```

```python
# Given two strings, check if one is a substring of the other without using the 'in' operator.
string_1 = input("Enter the string 1 : ")
string_2 = input("Enter the string 2 : ")
for i in range(len(string_1) - len(string_2) + 1):
    if string_1[i:i+len(string_2)] == string_2:
        print("Substring")
        break
else:
    print("Not a substring")
```

```python
# Reverse each word in a sentence while keeping the word order the same
string = input("Enter the string : ")
for i in string.split():
    print(i[::-1],end = " ")
```

```python
# Write a regex to extract all hashtags (e.g., #ai, #safety) from a text.
import re
text = input("Enter the string : ")
hashtags = re.findall(r"\w+", text)
print(hashtags)
```

```python
# Create a regex that validates whether a given input is a proper IPv4 address.
import re
ip = input("Enter IP address: ")
if re.match(r'^(\d{1,3}\.){3}\d{1,3}$', ip):
    print("Valid")
else:
    print("Invalid")
```

```python
# Using regex, extract all sequences of repeated characters (e.g., "aa", "bbb").
import re
text = input("Enter the string : ")
result = re.findall(r'(.)\1+', text)
print(result)
```

```python
# Remove all non-alphabetic characters from a string using regex.
import re
text = input("Enter the string: ")
cleaned = re.sub(r'[^A-Za-z]', '', text)
print(cleaned)
```

```python
# Split a paragraph into sentences based on '.', '?', and '!' using regex.
import re
text = input("Enter the paragraph: ")
sentences = re.split(r'[.?!]+', text)
sentences = [s.strip() for s in sentences if s.strip()]
print(sentences)
```

```python
# Check whether a string is a valid filename ending with .txt, .csv, or .json.
import re
filename = input("Enter the filename: ")
pattern = r'^.+\.(txt|csv|json)$'
if re.match(pattern, filename):
    print("Valid filename")
else:
    print("Invalid filename")
```

```python
# Write a script that masks all digits in a string with '*' using regex substitution.
import re
text = input("Enter the string: ")
masked = re.sub(r'\d', '*', text)
print(masked)
```

```python
# Extract all capitalized words (e.g., names) from a multiline text using regex.
import re
text = input("Enter the text:\n")
capitalized_words = re.findall(r'\b[A-Z][a-zA-Z]*\b', text)
print(capitalized_words)
```
