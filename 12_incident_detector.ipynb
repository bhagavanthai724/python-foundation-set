```python
# Build the core Incident Detector that reads the structured log output from Day 11,
# identifies events that exceed defined thresholds (error spikes, repeated failures, unusual patterns),
# assigns a severity tag (low/medium/high),
# and writes the detected incidents into a clean JSON file.
import json
from collections import Counter, defaultdict
log_file = "day11_structured.log"
output_file = "incidents.json"
events = []
with open(log_file) as f:
    for line in f:
        events.append(json.loads(line))
codes = [e["event"] for e in events]
counts = Counter(codes)
incidents = []
for ev in events:
    event = ev["event"]
    count = counts[event]
    if count > 20:
        severity = "high"
    elif count > 10:
        severity = "medium"
    elif count > 5:
        severity = "low"
    else:
        continue
    incidents.append({
        "timestamp": ev["timestamp"],
        "event": event,
        "count": count,
        "severity": severity
    })
with open(output_file, "w") as f:
    json.dump(incidents, f, indent=2)
```
