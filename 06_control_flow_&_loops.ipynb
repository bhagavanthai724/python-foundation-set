```python
# Using a while loop, keep taking user input until the user types "exit", then print how many inputs were entered.
count = 0
while True:
    user_input = input("Enter 'exit' to stop : ")
    if user_input.lower() == "exit":
        break
    count += 1
print("You entered", count, "inputs.")
```

```python
# Create a script that checks if a number is prime using a for loop.
num = int(input("Enter a number: "))
if num < 2:
    print(num, "is not a prime number.")
else:
    is_prime = True
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
           break
    if is_prime:
        print(num, "is a prime number.")
    else:
        print(num, "is not a prime number.")
```

```python
# Write a program that counts how many vowels are in a sentence using a loop but without using built-in count().
count = 0
sentence = input("Enter the string : ")
for i in sentence:
    if i in "AEIOUaeiou":
        count += 1
print(count)
```

```python
# Use a loop to calculate the factorial of a user-provided number.
num = int(input("Enter a number: "))
factorial = 1
for i in range(1, num + 1):
    factorial *= i
print(f"Factorial of {num} is {factorial}")
```

```python
# Create a menu-driven CLI with options: add number, remove number, show list, exit.
lst = []
while True:
    print("\n1.Add  2.Remove  3.Show  4.Exit")
    ch = input("Choice: ")
    if ch == "1":
        lst.append(input("Number: "))
    elif ch == "2":
        num = input("Number: ")
        lst.remove(num) if num in lst else print("Not found")
    elif ch == "3":
        print(lst)
    elif ch == "4":
        break
    else:
        print("Invalid")
```

```python
# Write a program that iterates over a dictionary and prints key, value, and data type of each value.
dct = {"A" : 1 , "B" : 2 , "C" : 3 , "D" : 4}
for key,value in dct.items():
    print(f"Key : {key} , Value : {value} , Type : {type(value)}")
```

```python
# Using nested loops, print a right-aligned triangle of stars (*) of height 10.
n = int(input("Enter the number : "))
for i in range(0, n + 1):
    for s in range(n - i):
        print(" ", end=" ")
    for j in range(i):
        print("*", end=" ")
    print()
```

```python
# Loop through a list of sentences and print the number of words in each sentence.
lst = ["I love Python", "Hello word", "AI is powerful"]
for sentence in lst:
    word_count = len(sentence.split())
    print(f"{sentence} : {word_count}")
```

```python
# Given a list of prices, use a loop to apply a 10% discount to items above 500.
prices = [120, 750, 300, 999, 450, 520]
for i in range(len(prices)):
    if prices[i] > 500:
        prices[i] = prices[i] * 0.9
print(prices)
```

```python
# Use a loop to merge two sorted lists into a new sorted list without using built-in sorting.
a = [1, 3, 5, 7]
b = [2, 4, 6, 8, 10]
i = j = 0
merged = []
while i < len(a) and j < len(b):
    if a[i] < b[j]:
        merged.append(a[i])
        i += 1
    else:
        merged.append(b[j])
        j += 1
while i < len(a):
    merged.append(a[i])
    i += 1
while j < len(b):
    merged.append(b[j])
    j += 1
print(merged)
```

```python
password = "1234"
attempts = 0
while attempts < 3:
    p = input("Enter password: ")
    if p == password:
        print("Login successful!")
        break
    else:
        print("Wrong password.")
        attempts += 1
if attempts == 3:
    print("Account locked. Too many attempts.")
```
