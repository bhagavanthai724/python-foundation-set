```python
# Write a script that converts temperature from Celsius to Fahrenheit based on user input.
celsius = int(input("Enter the celsius : "))
fahrenheit = (9/5) * celsius + 32
print(f"The Fahrenheit for given celsius {celsius} C is {fahrenheit} F")
```

```python
# Write a script to check if a number is divisible by 2, 3, or both.
n = int(input("Enter the number : "))
if n % 2 == 0 and n % 3 != 0:
    print("The given number is only divisible by 2 ")
elif n % 2 != 0 and n % 3 == 0:
    print("The given number is only divisible by 3 ")
elif n % 2 == 0 and n % 3 == 0:
    print("The given number is divisible by both 2 and 3 ")
else:
    print("Not divisible by 2 and 3")
```

```python
# Create a simple calculator program using basic operators (+, -, *, /).
a = int(input("Enter the initial number : "))
while True:
    b = int(input("Enter the next number : "))
    opt = input("Enter the operator (+,-,*,/) : ")
    if opt == "+":
        result = a + b
    elif opt == "-":
        result = a - b
    elif opt == "*":
        result = a * b
    elif opt == "/":
        result = a / b
    else:
        print("Invalid operator! Try again")
        continue
    print(f"Result = {result}")
    a = result
    inp = input("Do you want to continue (yes or no) : ").lower()
    if inp == "no":
        print("Thank you!")
        break
```

```python
# Write a Python script that compares two strings and prints whether they are equal (case-insensitive).
inp_1 = input("Enter the string 1 : ").lower()
inp_2 = input("Enter the string 2 : ").lower()
if inp_1 == inp_2 :
    print("Both strings are same")
else:
    print("Both strings are not same")
```

```python
# Create a script that asks the user for a sentence and prints the sentence in uppercase, lowercase, and titlecase.
s = input("Enter the sentence : ")
opt = input("Enter the operation (uppercase , lowercase , titlecase) : ").lower()
if opt == "uppercase":
    print(s.upper())
elif opt == "lowercase":
    print(s.lower())
elif opt == "titlecase":
    print(s.title())
else:
    print("Operator is invalid!")
```
