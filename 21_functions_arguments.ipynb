```python
# Write a function that takes two numbers and returns their sum.
def add_nums(a, b):
    return a + b

print(add_nums(3, 7))
```

```python
# Write a function that accepts a string and returns the number of vowels in it.
def count_vowels(s):
    vowels = "aeiouAEIOU"
    return sum(1 for ch in s if ch in vowels)

print(count_vowels("Hello World"))
```

```python
# Create a function with a default argument that greets a user by name.
def greet(name="Guest"):
    return f"Hello, {name}!"
print(greet())
print(greet("Alice"))
```

```python
# Variable-length arguments (*args) returning their product.
def product(*args):
    result = 1
    for n in args:
        result *= n
    return result
print(product(2,3,4))
```

```python
# Function that accepts keyword arguments (**kwargs) and prints key-value pairs.
def show_kwargs(**kwargs):
    for k, v in kwargs.items():
        print(k, "=", v)
show_kwargs(a=1, b=2, name="Test")
```

```python
# Function that takes a filename and returns how many lines it contains.
def count_lines(filename):
    with open(filename, "r") as f:
        return sum(1 for _ in f)
# Example (requires an actual file)
# print(count_lines("sample.txt"))
```

```python
# Normalize text: lowercase + strip whitespace.
def normalize(text):
    return text.strip().lower()
print(normalize("  HeLLo  "))
```

```python
# Function that takes two lists and returns list of paired tuples using zip().
def pair_lists(a, b):
    return list(zip(a, b))
print(pair_lists([1,2,3], ["a","b","c"]))
```

```python
# Check whether a number is prime.
def is_prime(n):
    if n < 2: return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True
print(is_prime(11))
```

```python
# Merge two dictionaries and return a new dictionary.
def merge_dicts(d1, d2):
    result = d1.copy()
    result.update(d2)
    return result
print(merge_dicts({"a":1}, {"b":2}))
```

```python
# Count frequency of each word in a sentence.
def word_frequency(sentence):
    words = sentence.lower().split()
    freq = {}
    for w in words:
        freq[w] = freq.get(w, 0) + 1
    return freq
print(word_frequency("hello world hello"))
```

```python
# Function that returns min, max, and average from list of numbers.
def stats(nums):
    return min(nums), max(nums), sum(nums)/len(nums)
print(stats([1,2,3,4,5]))
```

```python
# Validate email using regex.
import re
def valid_email(email):
    pattern = r"^[\w\.-]+@[\w\.-]+\.\w+$"
    return bool(re.match(pattern, email))
print(valid_email("test@example.com"))
```

```python
# Function that logs function calls: print the function name whenever it runs.
def log_call(func):
    def wrapper(*args, **kwargs):
        print(f"Running: {func.__name__}")
        return func(*args, **kwargs)
    return wrapper
@log_call
def sample():
    return "Done"
print(sample())
```
