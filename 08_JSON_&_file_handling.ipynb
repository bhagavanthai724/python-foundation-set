```python
# Write a script that opens a text file and prints the first 10 lines safely using a context manager.
with open("file.txt") as f:
    for i, line in enumerate(f):
        if i == 10: break
        print(line.rstrip())
```

```python
# Create a program that loads a JSON file and prints all top-level keys.
import json
data=json.load(open("data.json"))
print(list(data.keys()))
```

```python
# Write a utility that reads a JSON file and checks whether a given key exists.
import json, sys
d=json.load(open("data.json"))
print(sys.argv[1] in d)
```

```python
# Convert a Python dictionary into a JSON file with proper indentation.
import json
d={"a":1,"b":2}
json.dump(d, open("out.json","w"), indent=4)
```

```python
# Build a script that merges two JSON files into one combined JSON object.
import json
a=json.load(open("a.json"))
b=json.load(open("b.json"))
json.dump({**a,**b}, open("merged.json","w"), indent=4)
```

```python
# Write a program that reads a list from JSON and saves only unique items to a new JSON file.
import json
lst=json.load(open("list.json"))
json.dump(list(set(lst)), open("unique.json","w"), indent=4)
```

```python
# Create a file-copy tool that copies contents from source.txt to destination.txt.
open("destination.txt","w").write(open("source.txt").read())
```

```python
# Build a validator that checks if a JSON file is valid; catch JSONDecodeError if invalid.
import json
try:
    json.load(open("file.json"))
    print("valid")
except json.JSONDecodeError:
    print("invalid")
```

```python
# Write a script that reads a JSON list of objects and prints all objects where "status" == "active".
import json
for obj in json.load(open("items.json")):
    if obj.get("status")=="active":
        print(obj)
```

```python
# Create a program that loads a JSON configuration file and prints environment variables (dev/test/prod).
import json
cfg=json.load(open("config.json"))
print(cfg.get("dev"), cfg.get("test"), cfg.get("prod"))
```

```python
# Write a tool that backs up a file by creating a copy with a timestamp appended to the filename.
import time, shutil
ts=time.strftime("%Y%m%d_%H%M%S")
shutil.copy("file.txt", f"file_{ts}.txt")
```

```python
# Build a script that loads a nested JSON and prints all nested keys using recursion.
import json
def walk(d):
    if isinstance(d, dict):
        for k,v in d.items():
            print(k); walk(v)
    elif isinstance(d, list):
        for x in d: walk(x)
walk(json.load(open("nested.json")))
```

```python
# Write a program that appends a new entry to an existing JSON array and saves it back.
import json
arr=json.load(open("arr.json"))
arr.append({"new":"item"})
json.dump(arr, open("arr.json","w"), indent=4)
```

```python
# Create a utility that converts a .txt file into a JSON file where each line becomes a list element.
import json
lines=[l.rstrip() for l in open("file.txt")]
json.dump(lines, open("file.json","w"), indent=4)
```

```python
# Write a script that loads a JSON log file and filters all entries with severity >= "warning".
import json
levels=["info","debug","warning","error","critical"]
logs=json.load(open("log.json"))
for x in logs:
    if levels.index(x["severity"]) >= levels.index("warning"):
        print(x)
```
