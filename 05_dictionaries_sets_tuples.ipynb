```python
# Given a tuple of words, create a dictionary that maps each word to its length.
tup = ("AI", "Machine", "Learning", "Data")
dct = {}
for word in tup:
    dct[word] = len(word)
print(dct)
```

```python
# Create a dictionary of student names and marks, then print all students scoring above 75.
dct = {"Sundhar Pichai" : 76 , "Elon Musk" : 67 ,"Mark zuckerberg" : 89}
for name,mark in dct.items():
    if mark > 75:
        print(f"{name} : {mark}")
```

```python
# Take a sentence and build a dictionary that maps each word to its frequency.
sentence = "AI is the most powerful tool of modern world still fails to compete with OI , But AI dominate the current global roles"
words = sentence.split()
freq = {}
for word in words:
    if word in freq:
        freq[word] += 1
    else:
        freq[word] = 1
print(freq)
```

```python
# Merge two dictionaries into one without using the update() function.
dict1 = {"AI": 4 , "ML": 3 , "DL" : 2}
dict2 = {"GAI":1}
for k, v in dict1.items():
    dict2[k] = v
print(dict2)
```

```python
# Create a dictionary that maps numbers from 1 to 10 to their squares.
dct = {}
for i in range(1, 11):
    dct[i] = i ** 2
print(dct)
```

```python
# Given a dictionary, invert it so values become keys and keys become values.
# Given a dictionary, invert it so values become keys and keys become values.
dct = {"AI": 1, "ML": 2, "DL": 3}
inv = {}
for k, v in dct.items():
    inv[v] = k
print(inv)
```

```python
# Write a script to count how many unique characters exist in a string using a set.
string = "I love the puppy , because puppy is more cute"
s = set(string)
print(f"Count : {len(s)}")
```

```python
# Take a list with duplicates and convert it into a set, then print how many items were removed.
lst = [1,2,2,3,4,4,5]
s = set(lst)
count = len(lst) - len(s)
print(f"Count : {count}")
```

```python
# Given two sets of users (active and blocked), print only those who are active but not blocked.
active = {"AI", "ML", "DL", "NLP"}
blocked = {"DL", "NLP"}
result = active - blocked
print(result)
```

```python
# Create a tuple of 10 integers and print the max, min, and sum of the tuple.
tup = (1,2,3,4,5,6,7,8,8,0)
print(f"Max : {max(tup)} , Min : {min(tup)} , Sum : {sum(tup)}")
```

```python
# Given a tuple of filenames, extract only those ending with '.json' or '.csv'.
tup = ("AI.json","Python.csv","ML.py","DL.json")
for i in tup:
    if i.endswith((".json",".csv")):
        print(i)
```

```python
# Convert a tuple to a list, modify an element, and convert it back to a tuple.
tup = (0,2,3,4,5)
lst = list(tup)
lst[0] = 1
print(tuple(lst))
```

```python
# Using a set, find the intersection of two lists of IP addresses.
list1 = ["192.168.1.1", "10.0.0.5", "172.16.0.2"]
list2 = ["10.0.0.5", "192.168.1.1", "8.8.8.8"]
print(set(list1) & set(list2))
```

```python
# Create a dictionary from two lists: one for keys, one for values.
lst1 = ["AI", "ML", "DL", "GAI"]
lst2 = [4, 3, 2, 1]
dct = {}
for i in range(len(lst1)):
    dct[lst1[i]] = lst2[i]
print(dct)
```

```python
# Build a dictionary where keys are filenames and values are their file extensions.
files = ["AI.json", "model.py", "data.csv", "notes.txt"]
dct = {}
for f in files:
    name, ext = f.split(".")
    dct[name] = ext
print(dct)
```
